### np.subtract

```python
def np_subtract( a , b):
    # if both are plain numbers handle directly
    if not isinstance(a ,list) and not isinstance(b, list): # scalar
        if not isinstance(a,(int , float)) or not isinstance(b,(int,float)): # type check
            raise TypeError("input should be int or float")
        return a - b
    
    # if one element is list and another elements is not
    # broadcasting
    if isinstance(a , list) and not isinstance(b , list):
        result = []
        for x in a:
            result.append(np_subtract(x , b))
        return result
    if isinstance( b ,list) and not isinstance(a , list):
        result = []
        for y in b:
            result.append(np_subtract(y , a))
        return result

    # both a , b should have same length
    # both lists
    if isinstance(a, list) and isinstance(b, list):
        if len(a) == 0 and len(b) == 0:
            return []
        if len(a) == 0 or len(b) == 0:
            raise ValueError("both inputs must be non-empty!")
        if len(a) != len(b):
            raise ValueError("length of a must match length of b")
        result = []
        for i in range(len(a)):
            result.append(np_subtract(a[i], b[i]))
        return result
    
```
