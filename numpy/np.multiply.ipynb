### np.multiply

```python
def np_multiply(a , b):
    # if both are scalars handle directly
    if not isinstance(a,list) and not isinstance(b , list):
        if not isinstance(a,(int , float)) or not isinstance(b,(int,float)):
            raise TypeError("input must be int or float ")
        return a * b

    # if one is list and other is scalar
    # broadcasting
    if isinstance(a ,list) and not isinstance(b , list):
        result = []
        for x in a:
            result.append(np_multiply(x,b))  # recursion
        return result
    if isinstance(b , list) and not isinstance(a ,list):
        result = []
        for y in b:
            result.append(np_multiply(y,a))
        return result

    # if  both inputs are lists then they must be samelength
    if isinstance(a , list) and isinstance(b , list): # empty case
        if len(a) == 0 and len(b) == 0:
            return []
        if len(a) == 0 or len(b) == 0:
            raise ValueError("both inputs must be non-empty!")

        
        if len(a) != len(b):
            raise ValueError("length of inputs must be same")
        result = []
        for i in range(len(a)):
            result.append(np_multiply(a[i],b[i]))
        return result
    
```
