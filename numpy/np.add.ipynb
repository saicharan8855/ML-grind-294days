### np.add

```python
def np_add(a, b):

    # helper function inside np_add
    def is_scalar(x):
        return not isinstance(x, list)

    # scalar case
    if is_scalar(a) and is_scalar(b):
        return a + b

    # scalar broadcasting
    if is_scalar(a):
        return [np_add(a, x) for x in b]

    if is_scalar(b):
        return [np_add(x, b) for x in a]

    # shape mismatch
    if len(a) != len(b):
        raise ValueError(f"Shapes not compatible: {len(a)} vs {len(b)}")

    # recursive elementwise addition
    return [np_add(x, y) for x, y in zip(a, b)]
    
```
